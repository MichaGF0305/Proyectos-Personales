#### Librerías a utilizar

```python
import numpy as np
from scipy.stats import pearsonr, spearmanr, kendalltau
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

### Desarrollo de la función para los 3 ejercicios

```python
# Función general para análisis de regresión y cálculo de todas las correlaciones
def analisis_regresion(x, y):
    # Crear modelo de regresión lineal
    modelo = LinearRegression()
    modelo.fit(np.array(x).reshape(-1, 1), y)
    
    # Coeficientes de la regresión
    pendiente = modelo.coef_[0]
    interseccion = modelo.intercept_
    
    # Predicciones de Y
    y_pred = modelo.predict(np.array(x).reshape(-1, 1))
    
    # Calcular todas las correlaciones
    corr_pearson, _ = pearsonr(x, y)
    corr_spearman, _ = spearmanr(x, y)
    corr_kendall, _ = kendalltau(x, y)
    
    # Devolver los resultados
    return pendiente, interseccion, corr_pearson, corr_spearman, corr_kendall, y_pred
```

## Ejercicio 1: Velocidad Promedio vs. Número de Accidentes

**Datos:**

| Semana | Velocidad Promedio (km/h) | Número de Accidentes |
|--------|---------------------------|----------------------|
| 1      | 50                        | 5                    |
| 2      | 55                        | 7                    |
| 3      | 60                        | 6                    |
| 4      | 65                        | 8                    |
| 5      | 70                        | 10                   |
| 6      | 75                        | 9                    |

**Tareas:**
1. Calcular la ecuación de la línea de regresión lineal que predice el número de accidentes (Y) en función de la velocidad promedio (X).
2. Calcular el coeficiente de correlación para determinar la fuerza y la dirección de la relación entre la velocidad promedio y el número de accidentes.

---

```python
# Datos
velocidad_prom = [50, 55, 60, 65, 70, 75]
num_accidentes = [5, 7, 6, 8, 10, 9]
```

```python
pendiente_1, interseccion_1, corr_1, corr_spearman_1, corr_kendall_1, y_pred_1 = analisis_regresion(velocidad_prom, num_accidentes)

print(f"\nEjercicio 1 (Velocidad vs Accidentes):")
print(f"Pendiente: {pendiente_1:.2f}, Intersección: {interseccion_1:.2f}")
print(f"Correlación Pearson: {corr_1:.2f}")
print(f"Correlación Spearman: {corr_spearman_1:.2f}")
print(f"Correlación Kendall: {corr_kendall_1:.2f}")
print(f"Predicciones: {y_pred_1}")
```

```python
# Visualización de los resultados

plt.figure(figsize=(8, 6))
plt.scatter(velocidad_prom, num_accidentes, color='blue', label='Datos Reales')
plt.plot(velocidad_prom, y_pred_1, color='red', label='Regresión Lineal')
plt.xlabel('Velocidad Promedio (km/h)')
plt.ylabel('Número de Accidentes')
plt.legend
plt.show
```
